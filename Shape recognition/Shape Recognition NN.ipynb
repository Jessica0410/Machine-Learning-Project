```python
# image manipulation
from PIL import Image
```

```python
# label the data
def label(path):
    images, labels = [],[]
    l1 = os.listdir(path)
    for i in l1:
        l2 = os.listdir(path+'/'+i)
        for j in l2:
            image = Image.open(path+'/'+i+'/'+j)
            image.resize(size = (32,32))
            image.convert('L')
            images.append(np.array(image).flatten())
            labels.append(i)
    return np.array(images), labels
```

```python
X,y = label('shapes/shapes')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y)
y = le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
#Train test split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.8, stratify=y, shuffle = True)
#Crossvalidation
X_cv, X_test, y_cv, y_test = train_test_split(X_test,y_test,train_size = 0.5, stratify=y_test, shuffle = True)
```

```python
y_cv
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model =Sequential([
    tf.keras.Input(shape=(X_train.shape[1]),),   
    Dense(64, activation = 'relu', name = "L1"),
    Dense(32, activation = 'relu', name = "L2"), 
    Dense(3, activation = 'linear', name = "L5")   
], name = 'Shape_Recognition')
model.summary()
```

```python
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(0.001),
    metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs = 240, validation_data = (X_cv, y_cv))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
```

```python
pre = []
count = 0 # count misclassification
for i in range(len(X_test)):
    shape = X_test[i]
    prediction = model.predict(shape.reshape(1,2352))
    prediction_p = tf.nn.softmax(prediction)
    yhat = np.argmax(prediction_p)
    pre.append([y_test[i], yhat])
    if y_test[i]!=yhat:
        count+=1
pre
```
